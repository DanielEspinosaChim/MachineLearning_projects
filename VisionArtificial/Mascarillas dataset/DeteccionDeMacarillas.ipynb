```python
import cv2
import os
import numpy as np
```

```python
data_path = 'Dataset_faces'
```

```python
dir_list = os.listdir(data_path)
print("lista de directorios:", dir_list)
```

```python
labels = []
facesData = []
label = 0

for name_dir in dir_list:
    dir_path = data_path + '/' + name_dir
    for file_name in os.listdir(dir_path):
        img_path = dir_path + '/' + file_name
        print("img_path:", img_path)
        img = cv2.imread(img_path, 0)
        #cv2.imshow('image', img)
        #cv2.waitKey(10)
        
        facesData.append(img)
        labels.append(label)
        
    label = label + 1
        
    
print("Etiqueta 0", np.count_nonzero(np.array(labels) == 0))
print("Etiqueta 1", np.count_nonzero(np.array(labels) == 1))
```

```python
#train model
face_mask = cv2.face.LBPHFaceRecognizer_create()
print("Entrenando...")
face_mask.train(facesData, np.array(labels))
```

```python
face_mask.write('mask_model.xml')
print("Modelo almacenado...")
```
